```python
import turicreate as tc
```

## Load some house sales data

```python
sales=tc.SFrame('home_data.gl/')
```

```python
sales
```

## Exploring the data for housing sales

```python
sq=tc.SArray(sales["sqft_living"])
sq
p=tc.SArray(sales["price"])
p
```

```python
tc.visualization.scatter(sq,p,xlabel="SQ_ft",ylabel="Price",title="house")
```

## Create a simple regression model of sq_ft living to price

```python
train_data,test_data=sales.random_split(0.8,seed=0)
```

## Build the regression model


```python
sqft_model=tc.linear_regression.create(train_data,target='price',features=["sqft_living"])
```

### This is a regression model used by the turicreate create to learn from the training data.It uses the method called Newtons method.and this algorithm allows us to make some predictions on the data.Lets use this linear regression model to evaluate the prediction.

## Evaluate the simple model

```python
print (test_data["price"].mean())
```

```python
sqft_model.evaluate(test_data)
```

### Let's show what our predictions look like

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
plt.plot(test_data['sqft_living'],test_data['price'],'.',
         test_data['sqft_living'],sqft_model.predict(test_data),'-')
        
```

#### Dots represent the training data and the dashes represent the test data

```python
sqft_model.coefficients
```

## Build a regression model with more features 

```python
my_features_model=tc.linear_regression.create(train_data,target='price',features=my_features)
```

```python
print (my_features)
```

### Now we have two models first is the sq_ft model and the other one is my_features model,so lets compare their performance

```python
print(sqft_model.evaluate(test_data))
print(my_features_model.evaluate(test_data))
```
